```python
import yfinance as yf
```

```python
ticker=yf.Ticker("tsla")
tickerhist = ticker.history(start="1980-01-01", end="2023-05-15", interval='1wk')[['Open','Close']]
tickerhist['Open'].plot()
```

```python
#msfthist[0:3['Open','Close']]
percentchange=[]
dataLength=len(tickerhist)
for k in range(dataLength):
    percentchange.append((tickerhist.iloc[k,1]/tickerhist.iloc[k,0])-1)
tickerhist['Percentchange']=percentchange
tickerhist['Percentchange'].plot()
```
