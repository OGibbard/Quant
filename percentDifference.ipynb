```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
msft=yf.Ticker("msft")
msfthist = msft.history(start="1980-01-01", end="2023-05-15", interval='1d')[['Open','Close']]
msfthist['Open'].plot()
```

```python
#msfthist[0:3['Open','Close']]
percentchange=[]
dataLength=len(msfthist)
for k in range(dataLength):
    percentchange.append((msfthist.iloc[k,1]/msfthist.iloc[k,0])-1)
msfthist['Percentchange']=percentchange
msfthist['Percentchange'].plot()
```
