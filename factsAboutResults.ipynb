```python
import pickle
```

```python
# Load the file using pickle
with open("../Data/SP500/AAPL-minute2021-269kParams.pkl", "rb") as file:
    AAPLData = pickle.load(file)

with open("../Data/SP500/SPY-minute2021-269kParams.pkl", "rb") as file:
    SPYData = pickle.load(file)

#Sort results by each trade and by total
allResults1 = AAPLData
allResults2 = SPYData

#Minimum number of trades to be considered
minTotal1=10
maxTotal1=100000
minTotal2=10
maxTotal2=100000
usedResults1=[]
usedResults2=[]

for result in allResults1:
    if maxTotal1>=result[2]>=minTotal1:
        usedResults1.append(result)

for result in allResults2:
    if maxTotal2>=result[2]>=minTotal2:
        usedResults2.append(result)
    
```

```python
sortedResults1 = sorted(usedResults1.copy(), reverse=True)
sortedTotalResults1 = sorted(usedResults1.copy(), reverse=True, key=lambda x: x[1])

sortedResults2 = sorted(usedResults2.copy(), reverse=True)
sortedTotalResults2 = sorted(usedResults2.copy(), reverse=True, key=lambda x: x[1])
```

```python
for k in range(20):
    print(sortedResults1[k][0:3],sortedResults2[k][0:3])
```

```python
for k in range(20):
    print(sortedTotalResults1[k][0:3],sortedTotalResults2[k][0:3])
```
