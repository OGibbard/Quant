```python
import requests
from bs4 import BeautifulSoup

def ChangeDate(month, day, year):
    if day>1:
        day = day-1
    else:
        if month>1:
            day = 31
            month = month-1
        else:
            year = year -1
            day = 31
            month = 12
    return month, day, year

currentMonth = 5
currentDay = 15
currentYear = 2023
totalScrapedInfo=[]

#7198
while currentYear>2002:
    currentMonth, currentDay, currentYear = ChangeDate(currentMonth, currentDay, currentYear)
    firstPart='http://openinsider.com/screener?s=&o=&pl=&ph=&ll=&lh=&fd=-1&fdr='
    secondPart='&td=0&tdr=&fdlyl=&fdlyh=&daysago=&xp=1&xs=1&vl=&vh=&ocl=&och=&sic1=-1&sicl=100&sich=9999&grp=0&nfl=&nfh=&nil=&nih=&nol=&noh=&v2l=&v2h=&oc2l=&oc2h=&sortcol=0&cnt=1000&page=1'
    currentDate = (str(currentMonth)+'%2F'+str(currentDay)+'%2F'+str(currentYear)+'+-+'+str(currentMonth)+'%2F'+str(currentDay)+'%2F'+str(currentYear))
    link = firstPart+currentDate+secondPart
    page = requests.get(link) # Getting page HTML through request
    soup = BeautifulSoup(page.content, 'html.parser') # Parsing content using beautifulsoup
    data = soup.select("table.tinytable tr td") # Selecting all of the anchors with titles
    
    print(currentMonth,currentDay,currentYear)

    for anchor in data:
        scrapedInfo = {
            "title": anchor.text
        }
        totalScrapedInfo.append(scrapedInfo)
```

```python
k=0
columns=17
file = open('../data/insiderTradingDataFull1.csv', 'w')
for stock in totalScrapedInfo:
    k=k+1
    temp=(str(*stock.values())).replace(',', '')
    if temp!='':
        if 1<k<14:
            file.write(temp+',')
    if k==columns:
        file.write('\n')
        k=0
file.close()
print('done')
```
